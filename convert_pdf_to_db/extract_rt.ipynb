```python
import zipfile
import os
import re
import pandas as pd
```

```python
data_folder_path = "data/"
```

# extract zip files

```python
file_paths = []
for folder_path, _, files in os.walk(data_folder_path):
    for file_name in files:
        full_path = os.path.join(folder_path, file_name)
        if('_RT' in file_name):
            europa_number = re.search(r'(\d+)_RT', file_name).group(1)
            file_paths.append({'europa_number': europa_number,
                                  'file_path': full_path})
```

```python
# Code to unzip files 
def unzip_files(zip_path):
    file_path = zip_file["file_path"]
    print(f'unzipping {file_path}', end='\r')
    extract_to = os.path.join('data_out', f'{zip_file["europa_number"]}_RT')
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        try:
            zip_ref.extractall(extract_to)
        except Exception as e:
            print(f"Error : Couldn't extract {file_path}")

zip_files = [x for x in file_paths if x['file_path'].endswith('.zip')]
i = 0
for zip_file in zip_files:
    i+=1
    unzip_files(zip_file['file_path'])

print(f'\n {i} occurences')
```

# Extract data


## List all files

```python
file_paths = []

#get file that are at the root of each subfolder
for folder_path, folders, files in os.walk(data_folder_path):
    for file_name in files:
        full_path = os.path.join(folder_path, file_name)
        if('_RT' in file_name and not file_name.endswith('.zip')):
            europa_number = re.search(r'(\d+)_RT', file_name).group(1)
            file_paths.append({'europa_number': europa_number,
                                  'file_path': full_path})
            

# we ge te files within folders named 'RT'
def find_folders_with_name(startpath, folder_name):
    folders_found = []
    for root, dirs, files in os.walk(startpath):
        for dir_name in dirs:
            if dir_name == folder_name:
                europa_number = re.search(r'data/(\d+)', root).group(1)
                folders_found.append({'europa_number': europa_number,
                                      'folder_path': os.path.join(root, dir_name)})
    return folders_found

for RT_folder in find_folders_with_name(data_folder_path, 'RT'):
    europa_number = RT_folder['europa_number']
    folder_path = RT_folder['folder_path']
    for file in os.listdir(folder_path):
        file_paths.append({
            'europa_number': europa_number,
            'file_path': os.path.join(folder_path, file)
        })
```
